```python
#MengImport library yang dibutuhkan
import numpy as np
import pandas as pd
```

```python
#MENAMBAHKAN DATASET
# Data yang digunakan adalah data belanja Konsumen dimana terdapat jenis kelamin, umur, penghasilan dan pengeluaran

#load dataset menggunakan URL
df = pd.read_csv("https://raw.githubusercontent.com/yuliapratiwi169/Handling-Missing-Value-and-Outliers-/main/shopping_data_missingvalue.csv")
```

```python
#Menampilkan 5 record teratas dari dataset yang telah di load, menggunakan head()

df.head()
```

```python
# melihat informasi dataset. terlihat bahwa pada kolom Age dari 200 data terdapat 192 data yang non null, artinya 8 data lainnya bernilai null/NaN
df.info()
```

```python
#ini digunakan untuk mendeskripsikan kolom yang memiliki missing value yaitu "Age"

df['Age'].describe()
```

```python
#Mengidentifikasi Missing Value Pada dataset. Terlihat bahwa terdapat 8 missing value pada kolom Age

df.isnull().sum()
```

```python
#Mengganti Missing value menggunakan Mean (Metode Imputation)

mean = df["Age"].mean()
df["Age"] = df["Age"].replace(np.nan, mean)
df.head()

#terlihat bahwa kolom age ke 2 yang semula NaN berubah menjadi nilai mean yaitu 49.0
```

```python
#Mengganti Missing value menggunakan Mode (Metode Imputation)

df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df.head()

#terlihat bahwa kolom age ke 2 yang semula NaN berubah menjadi nilai mode yaitu 32.0
```

```python
#Mengganti Missing value menggunakan Median (Metode Imputation)

median = df["Age"].median()
df["Age"] = df["Age"].replace(np.nan, median)
df.head()

#terlihat bahwa kolom age ke 2 yang semula NaN berubah menjadi nilai median yaitu 37.5
```

```python
#Panjang data 

print("\nPanjang Data :" , df.shape)
```

```python
# Menghapus Missing Value Pada Dataset

df.dropna(inplace=True)
print(df.isnull().sum())

#setelah dihapus Missing Value pada kolom age menjadi 0
```
